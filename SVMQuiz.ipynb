## 12. สร้าง train/test split ของข้อมูล สามารถลองทดสอบ 70:30, 80:20, 90:10 ratio ได้ตามใจ

```python
from sklearn.model_selection import train_test_split
```

```python
# select independent and dependent features
X = df.drop(['diagnosis_result'], axis=1).copy()
y = df['diagnosis_result'].copy()
```

```python
X.head()
```

```python
# convert y to dummy variable
y = pd.get_dummies(y,drop_first=True)
y = np.array(y).reshape(-1,)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
print('X_train shape : ', X_train.shape)
print('y_train shape : ', y_train.shape)
print('X_test shape : ', X_test.shape)
print('y_test shape : ', y_test.shape)
```

## 14. พิจารณาว่าควรทำ Normalization หรือ Standardization หรือไม่ควรทั้งสองอย่าง พร้อมให้เหตุผล

```python
fig = plt.figure(figsize=(12,6))
i = 1
for col in df.columns :
    if col != 'diagnosis_result':
        plt.subplot(2,4,i)
        plt.hist(df[col], bins=10)
        plt.xlabel(col)
        i+=1
fig.tight_layout()
plt.show()
```

Note :  
* feature แต่ละตัว มีหน่วยที่แตกต่างกันมาก ตั้งแต่ ทศนิยม 2 ตำแหน่ง จนถึงหลักพัน จึงควรมีการทำ feature scaling
* จาก histogram ในข้อ 10 ทำให้เห็นว่า perimeter, smoothness และ symmetry มี distribution คล้ายกับ normal distribution ควรใช้ feature scaling ชนิด standardization (จะทำให้ได้ข้อมูลอยู่ในช่วง (-3,3))
* area, compactness และ fractal_dimension มีลักษณะ right-skewed ใช้ standardization ก็ไม่ทำให้ความเบ้ลดลง ดังนั้นจะใช้ standardization หรือ normalization ก็ได้ ในที่นี้ขอเลือกใช้ normalization
* radius และ texture ไม่เป็น normal distriution จึงเลือกใช้ normalization

## 16. วัดผลโมเดล โดยใช้ confusion matrix และ ประเมินผลด้วยคะแนน Accuracy, F1 score, Recall, Precision

```python
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# default
# create model, fit and predict
clf = SVC()
clf.fit(X_train, y_train)
y_pred1 = clf.predict(X_test)
# model evaluation
cm1 = metrics.confusion_matrix(y_test, y_pred1)
accuracy1 = metrics.accuracy_score(y_test, y_pred1)
f11 = metrics.f1_score(y_test, y_pred1)
precision1 = metrics.precision_score(y_test, y_pred1)
recall1 = metrics.recall_score(y_test, y_pred1)
print('Defalut')
print('Confusion Matrix : \n', cm1)
print('Accuracy : ', accuracy1)
print('F1 score : ', f11)
print('Precision : ', precision1)
print('Recall : ', recall1)
```

Note :  
ในกรณีนี้ y = 1 คือ Malignant หรือเป็นเนื้อร้าย  
ดังนั้นโมเดลจะต้องพยายามลด False Negative  
โดยที่ False Negative แสดงได้จากค่า Recall

```python
# standardization
std_X = StandardScaler()
std_X_train = std_X.fit_transform(X_train)
std_X_test = std_X.transform(X_test)
# create model, fit and predict
clf = SVC()
clf.fit(std_X_train, y_train)
y_pred2 = clf.predict(std_X_test)
# model evaluation
cm2 = metrics.confusion_matrix(y_test, y_pred2)
accuracy2 = metrics.accuracy_score(y_test, y_pred2)
f12 = metrics.f1_score(y_test, y_pred2)
precision2 = metrics.precision_score(y_test, y_pred2)
recall2 = metrics.recall_score(y_test, y_pred2)
print('Standardization')
print('Confusion Matrix : \n', cm2)
print('Accuracy : ', accuracy2)
print('F1 score : ', f12)
print('Precision : ', precision2)
print('Recall : ', recall2)
```

```python
# normalization (min-max scaling)
mms_X = MinMaxScaler()
nor_X_train = mms_X.fit_transform(X_train)
nor_X_test = mms_X.transform(X_test)
# create model, fit and predict
clf = SVC()
clf.fit(nor_X_train, y_train)
y_pred3 = clf.predict(nor_X_test)
# model evaluation
cm3 = metrics.confusion_matrix(y_test, y_pred3)
accuracy3 = metrics.accuracy_score(y_test, y_pred3)
f13 = metrics.f1_score(y_test, y_pred3)
precision3 = metrics.precision_score(y_test, y_pred3)
recall3 = metrics.recall_score(y_test, y_pred3)
print('Normalization')
print('Confusion Matrix : \n', cm3)
print('Accuracy : ', accuracy3)
print('F1 score : ', f13)
print('Precision : ', precision3)
print('Recall : ', recall3)
```

Note :  
Normalization ให้ผลลัพธ์ดีที่สุดในกรณีนี้

## 17. หาค่า parameter combination ที่ดีที่สุด สำหรับ Dataset นี้ โดยใช้ GridSearch (Hyperparameter Tuning)

```python
from sklearn.model_selection import GridSearchCV
```

```python
SVC()
```

```python
param_combination = {'C':[0.0001,0.0003,0.001,0.003,0.01,0.03,0.1,0.3,1,3,10,30,100,300,1000],
                    'gamma':[0.0001,0.0003,0.001,0.003,0.01,0.03,0.1,0.3,1,3,10,30,100,300,1000]}
```

```python
# grid_search กับ default X and y
grid_search = GridSearchCV(SVC(), param_combination, verbose=3)
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
y_pred4 = grid_search.predict(X_test)
# model evaluation
cm4 = metrics.confusion_matrix(y_test, y_pred4)
accuracy4 = metrics.accuracy_score(y_test, y_pred4)
f14 = metrics.f1_score(y_test, y_pred4)
precision4 = metrics.precision_score(y_test, y_pred4)
recall4 = metrics.recall_score(y_test, y_pred4)
print('Normalization')
print('Confusion Matrix : \n', cm4)
print('Accuracy : ', accuracy4)
print('F1 score : ', f14)
print('Precision : ', precision4)
print('Recall : ', recall4)
```

Note :  
เมื่อกำหนด parameter C=1, gamma=0.0001 fit และ evaluate กับ default X และ y จะได้ เมทริกซ์มีค่าน้อยลง
