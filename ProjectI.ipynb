1. จงเขียนโปรแกรมรับค่า Input จากผู้ใช้ ถ้าผู้ใช้ใส่ 0.0 จึงจะสิ้นสุดโปรแกรม คำนวณหาค่าเฉลี่ยของเลขทั้งหมดที่ผู้ใช้ใส่     และปริ้นค่าเฉลี่ยคูณด้วยจำนวนครั้งที่ใส่ทั้งหมด f

```python
total = 0
count = 0

while True :
    x = int(input('กรุณาใส่ตัวเลขที่ต้องการหาค่าเฉลี่ย หรือใส่ 0 เมื่อต้องการสิ้นสุดโปรแกรม : '))
    if x == 0 :
        break
    else :
        count += 1
        total += x

average = total / count

print('ค่าเฉลี่ยคูณด้วยจำนวนครั้งที่ใส่ข้อมูลทั้งหมด {} * {} = {}'.format(average, count, average * count))
```

2. จงเขียนโปรแกรมให้ผู้ใช้ใส่เลข ขอบเขตล่าง และ ขอบเขตบน และคำนวณหาเลขจำนวนเฉพาะ (Prime Number) ที่อยู่ระหว่างขอบเขตล่างและขอบเขตบน

```python
# function isprime
def isprime(n):
    if (n == 1):
        return False
    elif (n == 2):
        return True;
    else:
        for i in range(2, n):
            if(n % i == 0):
                return False
        return True             

while True :
    lower_bound = int(input('กรุณาใส่ ขอบเขตล่าง : '))
    upper_bound = int(input('กรุณาใส่ ขอบเขตบน (inclusive) : '))
    if upper_bound <= 0 or lower_bound <= 0 :
        print('\n ***ขอบเขตบนและล่างต้องเป็นเลขจำนวนเต็มบวกเท่านั้น***')
        continue
    elif upper_bound <= lower_bound :
        print('\n ***กรุณากำหนดขอบเขตใหม่ให้ถูกต้อง ขอบเขตล่างต้องน้อยกว่าขอบเขตบน***')
        continue
    else :
        # หา prime number ด้วยฟังก์ชัน isprime
        your_prime = [num for num in range(lower_bound, upper_bound + 1) if isprime(num) == True]
        break
print('จำนวนเฉพาะในช่วงที่กำหนด : {}'.format(your_prime))
```

18. สร้าง column ใหม่ โดยให้บรรจุ Time of the week (วันจันทร์ อังคาร....) โดยใช้ข้อมูลจาก last_review (Hint: ใช้ timestamp เข้ามาช่วย)

```python
# แปลงคอลัมน์ last_review เป็น datetime
df1['last_review'] = pd.to_datetime(df1['last_review'])
```

```python
df1.info()
```

```python
# หา weekday index
df1['weekday_index'] = df1.apply(lambda x : x['last_review'].weekday(), axis = 1)
df1.head()
```

```python
# หา Time of the week (วันจันทร์ วันอังคาร ...) จาก weekday index
def get_day_name(index):
    if index == 0 :
        return 'วันจันทร์'
    elif index == 1 :
        return 'วันอังคาร'
    elif index == 2 :
        return 'วันพุธ'
    elif index == 3 :
        return 'วันพฤหัสบดี'
    elif index == 4 :
        return 'วันศุกร์'
    elif index == 5 :
        return 'วันเสาร์'
    else :
        return 'วันอาทิตย์'

df1['Time_of_the_week'] = df1.apply(lambda x : get_day_name(x['weekday_index']), axis = 1)
df1.head()
```

18. สร้างคอลัมน์ใหม่ โดยให้บรรจุ Day-of-the-week (Mon, Tue....) โดยใช้ข้อมูลจาก last_review (Hint: ใช้ datetime กับ map มาช่วย) 

```python
# หา day-of-the-week
df1['Day_of_the_week'] = df1['last_review'].map(lambda x : x.strftime('%A'))
df1.head()
```
