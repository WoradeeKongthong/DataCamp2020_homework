```python
print('ใส่ความยาวของด้านแต่ละด้านของรูปสามเหลี่ยม')
x1 = float(input('ความยาวด้านที่ 1 : '))
x2 = float(input('ความยาวด้านที่ 2 : '))
x3 = float(input('ความยาวด้านที่ 3 : '))
# c คือ ด้านที่ยาวที่สุด
# a, b คือ ด้านที่เหลืออีกสองด้าน
y = [x1,x2,x3]
y.sort(reverse=True)
c = y[0]; a = y[1]; b = y[2]
if a + b <= c:
    print('ค่าที่กำหนด ไม่สามารถสร้างสามเหลี่ยมได้')
elif c ** 2 == a **2 + b ** 2 :
    print('ค่าที่กำหนด สามารถสร้างสามเหลี่ยมมุมฉากได้')
else :
    print('ค่าที่กำหนด สามารถสร้างสามเหลี่ยมได้ แต่ไม่ใช่สามเหลี่ยมมุมฉาก')
```

15. โรงละครสัตว์มีนก กับ วัว และต้องจ่ายภาษีสัตว์ 2 ขา 150 บาทต่อตัว   จ่ายภาษีสัตว์ 4 ขา 220 บาทต่อตัว เขียนโปรแกรมรับค่า input สองตัว ตัวแรกเป็นผลรวมจำนวนหัวของสัตว์ ตัวสองเป็นผลรวมจำนวนขาของสัตว์ จงหาว่าโรงละครสัตว์ต้องจ่ายภาษีเท่าใด

ให้ จำนวน x = จำนวนนก,y = จำนวนวัว 

ให้ h = จำนวนหัว, l = จำนวนขา

จะได้สมการ

               x + y      = h                  --1
           
               2*x + 4*y  = l                  --2
           
    จาก 1-->   2*x + 2*y  = 2*h                --3

    2-3  -->   2*y  = l - 2*h

               y  = (l - 2*h)/2                --4
                   
    จาก 1 -->  x  = h - (l - 2*h)/2

               x  = (4*h - l)/2                --5

```python
while True :
    print('ใส่จำนวน หัว และ ขา ของสัตว์ ให้ถูกต้อง')
    h = int(input('ผลรวมจำนวนหัวของสัตว์ '))
    l = int(input('ผลรวมจำนวนขาของสัตว์ '))
    y = (l - 2*h)/2
    x = (4*h - l)/2
    if x%1 == 0 and y%1 == 0 :
        break
        
print('จำนวน นก = ', x, 'ตัว')
print('จำนวน วัว = ', y, 'ตัว')
animalFee = x*150 + y*220
print('โรงละครสัตว์ต้องจ่ายภาษี = ', animalFee)
```
