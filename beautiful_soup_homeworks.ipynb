# Homework 1 : หาราคาทอง
ทำการใช้ beautifulsoup ไปอ่านค่าที่ https://www.goldtraders.or.th/ แล้วเอาราคาทองออกมา (print ออกมา หรือ save ลง file)

## 1.access the website

```
response = requests.get('https://www.goldtraders.or.th/default.aspx')
```

```
# check if the web page exist
response.status_code
```

```
# print the output
response.text
```

```
# extract content of the page
src = response.content
```

## 2.create a soup

```
soup = BeautifulSoup(src, 'lxml')
```

```
print(soup.prettify)
```

```
# get the div tag that contain the gold price to see the detail
details = soup.findAll(attrs={'id':'DetailPlace_uc_goldprices1_GoldPricesUpdatePanel'})
```

```
len(details)
```

```
print(details[0])
```

```
# วันที่ และ เวลา บันทึกราคาทองคำ
report_time = details[0].find(attrs={'id':'DetailPlace_uc_goldprices1_lblAsTime'}).string
print(report_time)
```

```
# ราคาทองคำแท่ง(bar) 96.5% ขายออก 
price1 = details[0].find(attrs={'id':'DetailPlace_uc_goldprices1_lblBLSell'}).string
price1 = str(price1).replace(',','')
print(price1)
```

```
# ราคาทองคำแท่ง(bar) 96.5% รับซื้อ
price2 = details[0].find(attrs={'id':'DetailPlace_uc_goldprices1_lblBLBuy'}).string
price2 = str(price2).replace(',','')
print(price2)
```

```
# ราคาทองรูปพรรณ(ornament) 96.5% ขายออก
price3 = details[0].find(attrs={'id':'DetailPlace_uc_goldprices1_lblOMSell'}).string
price3 = str(price3).replace(',','')
print(price3)
```

```
# ราคาทองรูปพรรณ(ornament) 96.5% ฐานภาษี
price4 = details[0].find(attrs={'id':'DetailPlace_uc_goldprices1_lblOMBuy'}).string
price4 = str(price4).replace(',','')
print(price4)
```

## 3.save to csv file

```
f = open('gold_price.csv','a')

f.write("report_time, bar_sell, bar_buy, ornament_sell, ornament_buy\n")
f.write(report_time + "," + price1 + "," + price2 + "," + price3 + "," + str(price4) + "\n")

f.close()
```

# Homework 2 : ราคาหุ้น
- ไปที่ https://www.settrade.com/C13_MarketSummary.jsp?detail=SET50
- อ่านค่าตารางข้างล่างทั้งหมด เอาข้อมูลมาเก็บไว้ (print ออกมา หรือ save ลง file)

## 1.access the website

```
response = requests.get('https://www.settrade.com/C13_MarketSummary.jsp?detail=SET50')
```

```
# confirm that the website exist
response.status_code
```

```
# get the source
src = response.content
print(src)
```

## 2.create a soup

```
soup = BeautifulSoup(src, 'lxml')
```

```
# print the soup
print(soup.prettify)
```

```
# get all tables from the page
tables = soup.find_all(attrs={'class':'table table-info'})
```

```
# our table is the first one
tables[0]
```

```
summary_table = tables[0]
```

```
# get all links in the table
links = summary_table.find_all('a')
```

```
# print string in each link
for link in links:
  print(link.string.rsplit()[0])
```

```
# store the string in variable
stock_index = [link.string.rsplit()[0] for link in links]
print(stock_index)
```

```
# find numerical data in the table
table_data = summary_table.find_all('td')
```

```
# print the result along with its index
for index, data in enumerate(table_data):
  print(index, " : ", data.string)
```

```
# store each column in a variable
recent = []
change = []
perc_change = []
max = []
min = []
volume_K = []
price_M = []

for i, data in enumerate(table_data) :
  if i%8 == 1 :
    recent.append(float(str(data.string).replace(',','')))
  if i%8 == 2 :
    change.append(float(str(data.string).replace(',','')))
  if i%8 == 3 :
    perc_change.append(float(str(data.string).replace(',','')))
  if i%8 == 4 :
    max.append(float(str(data.string).replace(',','')))
  if i%8 == 5 :
    min.append(float(str(data.string).replace(',','')))
  if i%8 == 6 :
    volume_K.append(float(str(data.string).replace(',','')))
  if i%8 == 7 :
    price_M.append(float(str(data.string).replace(',','')))
```

```
# print all of out results
print('stock_index : ', stock_index)
print('recent : ', recent)
print('change : ', change)
print('perc_change : ', perc_change)
print('min : ', min)
print('max : ', max)
print('volume_K : ',volume_K)
print('price_M : ', price_M)
```

## 3.store the result into dataframe and save to csv file

```
import pandas as pd
```

```
# store in dataframe
stock_df = pd.DataFrame(index=stock_index)
stock_df['recent'] = recent
stock_df['change'] = change
stock_df['perc_change'] = perc_change
stock_df['min'] = min
stock_df['max'] = max
stock_df['volume_K'] = volume_K
stock_df['price_M'] = price_M
stock_df
```

```
# save dataframe to csv file
stock_df.to_csv('stock_index_summary.csv')
```
